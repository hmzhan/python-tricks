### Python trick #1: enumerate

```python
x = [3, 3, 1, 5, 6, 7, 8]
for i, v in enumerate(x):
    print(i, v)
```

### python trick #2: .get()

```python
d = {
    'a': 1,
    'b': 2,
    'c': 3
}

print(d.get('a', 0))
print(d.get('d', 0))
```

### python trick #3: multi-line string

```python
test = """It is the simplest method to let a long string split into different lines. 
You will need to enclose it with a pair of Triple quotes, 
one at the start and second in the end."""
print(test)
```

### python trick #4: sort dictionary by value

```python
x = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
dict(sorted(x.items(), key=lambda item: item[1]))
```

```python
{k:v for k,v in sorted(x.items(), key=lambda item: item[1])}
```
