### Python trick #1: enumerate

```python
x = [3, 3, 1, 5, 6, 7, 8]
for i, v in enumerate(x):
    print(i, v)
```

### python trick #2: .get()

```python
d = {
    'a': 1,
    'b': 2,
    'c': 3
}

print(d.get('a', 0))
print(d.get('d', 0))
```

### python trick #3: multi-line string

```python
test = """It is the simplest method to let a long string split into different lines. 
You will need to enclose it with a pair of Triple quotes, 
one at the start and second in the end."""
print(test)
```

### python trick #4: sort dictionary by value

```python
x = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
dict(sorted(x.items(), key=lambda item: item[1]))
```

```python
{k:v for k,v in sorted(x.items(), key=lambda item: item[1])}
```

### python trick #5: merge dictionaries

```python
x = {'a': 1, 'b': 2}
y = {'d': 3, 'c': 4, 'b': 6}

z = {**x, **y}
print(z)
```

### python trick #6: transpose a matrix

```python
mat = [[8, 9, 10], [11, 12, 13], [0, 6, 8]]
new_mat=zip(*mat)
for row in new_mat:
    print(row)
```

### python trick #7: flatten a list

```python
import itertools
a = [[1, 2], [3, 4], [5, 6]]
b = list(itertools.chain.from_iterable(a))
print(b)
```

### python trick #8: f-string

```python
str_val = 'apples'
num_val = 42

print(f'{num_val} {str_val}') # 42 apples
```

```python
price_val = 6.12658

print(f'{price_val:.2f}')
```

```python
str_val = 'apples'

print(f'{str_val!r}')
```

```python
str_val = 'apples'
num_val = 42

print(f'{str_val=}, {num_val = }') # only works in python 3.8+
```

```python
from datetime import datetime;

date_val = datetime.utcnow()
print(f'{date_val=:%Y-%m-%d}') # only works in python 3.8+
```
